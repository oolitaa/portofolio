## Import Library

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
```

## Data Preprocessing

### Import Data

```python
automobile = pd.read_csv('Automobile_data.csv')
automobile.head()
```

### Finding Missing Value

```python
automobile.info()
```

```python
automobile.value_counts('normalized-losses').head()
```

#### Convert '?' to NaN

```python
automobile.replace("?", np.nan, inplace=True)
automobile.head()
```

```python
print(automobile.isnull().sum().where(lambda x: x > 0).dropna())
```

### Dealing with Missing Data 

1. Drop Data
   - drop the whole row -- Less than 5% missing values in rows
   - drop the whole column -- More than 50% missing values in a column
2. Replace Data
   - replace it by mean -- Numerical data, normally distributed
   - replace it by frequency -- Numerical data, skewed distribution or outliers
   - replace it based on other functions

#### Drop the whole row

```python
automobile.price.isnull().sum()
```

```python
automobile.dropna(subset=["price"], axis=0, inplace=True)
```

```python
# Reset Index 
automobile.reset_index(drop=True, inplace=True)
automobile.head()
```

#### Replace NaN by mean value 

##### normalized-losses

```python
avg_normalized = automobile['normalized-losses'].astype('float').mean(axis=0) ## axis 0 for column, 1 for row
avg_normalized
```

```python
automobile.replace({'normalized-losses': np.nan}, avg_normalized, inplace=True)
automobile.head()
```

##### bore

```python
automobile.bore.value_counts(dropna=False)
```

```python
avg_bore = automobile.bore.astype('float').mean(axis=0)
avg_bore
```

```python
automobile.replace({'bore':np.nan}, avg_bore, inplace=True)
```

```python
automobile.bore.isnull().sum()
```

##### stroke

```python
automobile.stroke.value_counts(dropna = False)
```

```python
avg_stroke = automobile.stroke.astype('float').mean(axis=0)
avg_stroke
```

```python
automobile.replace({'stroke':np.nan}, avg_bore, inplace=True)
```

```python
automobile.stroke.isnull().sum()
```

##### horsepower

```python
automobile.horsepower.value_counts(dropna = False)
```

```python
avg_horse = automobile.horsepower.astype('float').mean(axis = 0)
avg_horse
```

```python
automobile.replace({'horsepower':np.nan}, avg_horse, inplace=True)
```

```python
automobile.horsepower.isnull().sum()
```

##### peak-rpm

```python
automobile['peak-rpm'].value_counts(dropna= False)
```

```python
avg_peak = automobile['peak-rpm'].astype('float').mean(axis=0)
avg_peak
```

```python
automobile.replace({'peak-rpm':np.nan}, avg_peak, inplace=True)
```

```python
automobile['peak-rpm'].isnull().sum()
```

#### Replace NaN by frequency 

```python
automobile['num-of-doors'].value_counts(dropna=False)#idxmax()
```

```python
automobile.replace({'num-of-doors': np.nan}, 'four', inplace=True)
```

```python
automobile['num-of-doors'].value_counts(dropna=False)#idxmax()
```

### Data Formating

```python
automobile.dtypes
```

```python
# Numeric to Float 
automobile[["bore", "stroke", "price", "peak-rpm"]] = automobile[["bore", "stroke", "price", "peak-rpm"]].astype(float)
```

### Exploratory Data Analysis (EDA)

```python
automobile.head()
```

#### Data Normalization

```python
automobile.describe()
```

```python
scaler = MinMaxScaler()
cols_to_normalize = ["curb-weight", "engine-size", "compression-ratio", "peak-rpm", "wheel-base", "length", "width", "height"]
# transformation
automobile[cols_to_normalize] = scaler.fit_transform(automobile[cols_to_normalize])
automobile.head()
```

```python
automobile['horsepower'] = automobile['horsepower'].copy().astype(int)
```

```python
plt.hist(automobile['horsepower'], bins=10, edgecolor="black")

# Labels
plt.xlabel('Horsepower')
plt.ylabel('Count')
plt.title('Horsepower Bins')

plt.show()
```

#### Bining

```python
bins = np.linspace(automobile["horsepower"].min(), automobile["horsepower"].max(), 4)  # 4 bounderies â†’ 3 category
labels = ["Low", "Medium", "High"]
```

```python
automobile["horsepower_category"] = pd.cut(automobile["horsepower"], bins=bins, labels=labels, include_lowest=True)
```

```python
automobile[['horsepower', 'horsepower_category']].head()
```

```python
automobile['horsepower_category'].value_counts()
```

```python
plt.bar(automobile["horsepower_category"].value_counts().index, automobile["horsepower_category"].value_counts().values)

# Labels
plt.xlabel("Horsepower Category")
plt.ylabel("Count")
plt.title("Distribution of Horsepower Categories")

plt.show()
```
